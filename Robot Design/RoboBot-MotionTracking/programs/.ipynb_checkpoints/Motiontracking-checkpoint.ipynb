```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 

```

```python
def wall_detection(transform):
    hMin = 15
    sMin = 0
    vMin = 220
    hMax = 75
    sMax = 255
    vMax = 255

    transform_maze = transform.copy() 
    imgHSV = cv2.cvtColor(transform_maze, cv2.COLOR_BGR2HSV)

    lower = np.array([hMin, sMin, vMin])
    upper = np.array([hMax, sMax, vMax])
    mask = cv2.inRange(imgHSV, lower, upper)
    masking = cv2.bitwise_and(transform_maze, transform_maze, mask=mask)

    kernel = np.ones((3,3), np.uint8)
    opening = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)
    _,contours,_ = cv2.findContours(opening, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    contour = []
    for _,c in enumerate(contours):
            area = cv2.contourArea(c)
            if (area > 100):
                contour.append(c)          
    
    


    
    
    cv2.drawContours(transform_maze,contour, -1, (245, 0, 0), thickness=cv2.FILLED)
    transform_maze_RGB = cv2.cvtColor(transform_maze, cv2.COLOR_BGR2RGB)

    
    return contour, transform_maze;
```

```python
def robot_detector(maze):
    #hMin = 23
    #sMin = 45
    #vMin = 40
    hMin = 40
    sMin = 70
    vMin = 70
    hMax = 85
    sMax = 175
    vMax = 255
    maze_copy = maze.copy() 
    imgHSV = cv2.cvtColor(maze_copy, cv2.COLOR_BGR2HSV)

    lower = np.array([hMin, sMin, vMin])
    upper = np.array([hMax, sMax, vMax])
    mask = cv2.inRange(imgHSV, lower, upper)
    masking = cv2.bitwise_and(maze_copy, maze_copy, mask=mask)

    kernel = np.ones((5,5), np.uint8)
    opening = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)
    _,contours,_ = cv2.findContours(opening, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    contour = []
    for _,c in enumerate(contours):
            area = cv2.contourArea(c)
            if (area > 200):
                contour.append(c)
                (x, y), radius = cv2.minEnclosingCircle(c)
                center = (int(x), int(y))
                radius = int(radius)
    




    
    maze_copy_RGB = cv2.circle(maze, center, radius, (0, 0, 255), 2)
    maze_copy_RGB = cv2.cvtColor(maze_copy, cv2.COLOR_BGR2RGB)

    return center, radius
```

```python
def get_coords(center):
    if center[0] < 250:
        col = 0
    elif center[0] < 470:
        col = 1
    elif center[0] < 710:
        col = 2
    elif center[0] < 920:
        col = 3
    elif center[0] < 1100:
        col = 4
    elif center[0] < 1300:
        col = 5
    elif center[0] < 1500:
        col = 6
    elif center[0] < 1750:
        col = 7
    elif center[0] < 1920:
        col = 8
        
        
        
    if center[1] < 200:
        row = 0
    elif center[1] < 400:
        row = 1
    elif center[1] < 610:
        row = 2
    elif center[1] < 820:
        row = 3
    elif center[1] < 1100:
        row = 4
    elif center[1] < 1080:
        row = 5
        
    return row,col
```

```python
cap=cv2.VideoCapture("../mazevid.mp4")
file = "../position"
ret,frame=cap.read()
#1920x1080
position = set()
n = 0
while ret==True:
    key=cv2.waitKey(1)
    if key==ord('q') or ret == False:
        break
    ret,frame=cap.read()
    center, radius = robot_detector(frame)
    c,map_build = wall_detection(frame)
    coords = get_coords(center)
    position.add("In:(" + str(coords[0]) + "," + str(coords[1]) +")")
    cv2.putText(frame,"Row:" + str(coords[0]) + ", Col:" + str(coords[1]), (100,50), cv2.FONT_HERSHEY_SIMPLEX, 0.75, (50,170,50),2);
    cv2.imshow("Tracking", frame)

cv2.destroyAllWindows()
return;
```
